```python
"""

Docstring:

    I)This game is based on the first movie of the Harry Potter saga:
    The Sorcerer's Stone (2001). The user will play the role of Harry Potter, 
    in his quest of the Sorcerer's Stone. Each room the young wizard went through
    is represented by a stage in this game:
    
        1. Room 1: Decide how to get rid of the Devil's Snare (plant)
        2. Room 2: Decide how to get the flying key
        3. Room 3: Decide whether to play chess or not
        4. Room 4: Decide what to do facing the final opponent
        
    It is designed not to faithfully follow the original scenario but to provide
    the user with different ouputs. However, knowledge from the movie will be 
    useful.
    
    II) Known bugs and/or errors:
        None

"""

#Defining the first variable start to launch the game
def start():
    print("""
    You embody Harry Potter in his first year at Hogwarts. With your friends
    Hermione and Ron, you realize a malefic power is after a dangerous artefact:
    the sorcerer's stone. Despite all the warnings of the professors not to do 
    anything about it, you decide to investigate what is going on. Your inquiry
    soon leads you to a trap door garded by a three-headed dog. Luckily for you, 
    you manage to escape it by diving in the hatch.
    \n""")
    input(prompt = "<Press Enter to continue> \n")
    bad_plant()
    
#Defining the second variable bad_plant to launch the first stage
def bad_plant():
    print("""
    You landed on a massive slimmy plant. It immediately tries to strangle
    you. The only exit is by going through. What do you do young wizard?
    
    1. Use your wand to burn the plant
    2. Take the swiss knife in your pocket and cut its root
    3. You have already seen too much stuff and stay still while accepting your 
       fate
    \n""")    

#Prompting the user for a choice
    choice_1 = input(prompt = "Your choice \n\n")

#Defining simple conditional statement on strings to cover all possible answers
    if choice_1.lower() == '1' or 'wand' in choice_1.lower() or 'burn' in choice_1.lower():
        print("""
        Too bad you don't know the spell to light it up... Let's try again. 
        \n""")
        input(prompt = "<Press Enter to continue> \n")
        bad_plant() #Sends back the user to the beginning of this stage
        
    elif choice_1.lower() == '2' or 'knife' in choice_1.lower() or 'pocket' in choice_1.lower() or 'root' in choice_1.lower():
        print("""
        You are a wizard, why would you have a swiss knife on you? Let's try again.
        \n""")
        input(prompt = "<Press Enter to continue> \n")
        bad_plant()
        
    elif choice_1.lower() == '3' or 'still' in choice_1.lower() or 'move' in choice_1.lower() or 'stuff' in choice_1.lower():
        print("""
        Congratulations! You got through the plant and are know ready to 
        move on to the next stage.
        \n""")
        input(prompt = "<Press Enter to continue> \n")
        broom_stick() #Sends the user to the next variable broom_stick

#Else argument covers an unexpected answer from the user
    else:
        print("\n Come on, stick to the choice you little mudblood! \n")
        input(prompt = "<Press Enter to continue> \n")
        bad_plant()
    
#Defining the third variable broom_stick to launch the second stage    
def broom_stick():
    print("""
    You arrive in a dark room full of flying keys. To open the nex door, you will
    need only one of the key. To reach them, an old broomstick is available. 
    What do you do?
    
    1. Use the old broomstick
    2. Call for your nimbus 2000
    3. Cast a spell
    \n""")
    
    choice_2 = input(prompt = 'Your choice \n\n')
    
    if choice_2.lower() == '1' or 'broomstick' in choice_2.lower() or 'old' in choice_2.lower():
        print("""
        This broomstick is not fast enough to catch up with the keys!
        \n""")
        input(prompt = '<Press Enter to continue> \n')
        broom_stick()
    
    elif choice_2.lower() == '2' or 'call' in choice_2.lower() or 'nimbus' in choice_2.lower() or '2000' in choice_2.lower():
        print("""
        Unfortunately, you broke it in the last Quidditch game...
        \n""")
        input(prompt = '<Press Enter to continue> \n')
        broom_stick()
        
    elif choice_2.lower() == '3' or 'cast' in choice_2.lower() or 'spell'in choice_2.lower() or 'wand' in choice_2.lower():
        print("""
        Interesting... Among the 3 spells you have learned so far, what spell would 
        you like to cast? 
        
        1. Expelliarmus
        2. Immobulus
        3. Wingardium Leviosaaaaaa
        \n""")
        
        spell_lst = ['1', '2', '3']
        
#Creating a loop that will keep the user within as long as the answer is not correct
        while spell_lst != '2' or spell_lst != 'immobulus':
            spell = input(prompt = 'Your spell \n\n')
            spell = str(spell)
        
            if spell.lower() == '1' or 'expelliarmus' in spell.lower():
                print("""
                The spell bounced on the keys and knocked out Ron! While 
                he is uncounscious, let's try something else.\n
                """)
                            
            elif spell.lower() == '2' or 'immobulus' in spell.lower():
                print("""
                It was a fine spell! You now have time to catch the frozen key and
                open the door to the next room! \n
                """)
                chess_room() #Sends the user to the nex variable chess_room
                break #Break statement prevents the loop from continuing after the answer was found
                    
            elif spell.lower() == '3' or 'wingardium leviosa' in spell.lower():
                print("""
                It did not work. You only managed to get Hermione upset and is
                know correcting your pronounciation: it's Leviooooosa!
                """)
                
            else:
                print("""
                Incorrect entry. You should have stayed at the Dursleys.
                Try again.
                \n""") 
                
                
    else:
        print('\n Wrong entry, focus you muggle! \n')
        input(prompt = '<Press Enter to continue> \n')
        broom_stick()
        
        
#Defining the fourth variable chess_room to launch the third stage
def chess_room():
    print("""
    You enter and face a human size chess board game in the middle. Around
    the board, you notice a lot of rubbles. The next door is located on the other
    side of the chess board. Once again, you have to choose how to reach it:
    
    1. Cross the chess board 
    2. Go around the chess board
    3. Play chess
    4. This has gone too far, let's head back!
    \n""")
    
    choice_3 = input(prompt = 'Your choice \n\n')
    
    if choice_3.lower() == '1' or 'cross' in choice_3.lower():
        print("""
        The pawns are not nice enough to let you cross the board like that!
        Try something else.
        """)
        input(prompt = '<Press Enter to continue> \n')
        chess_room()
        
    elif choice_3.lower() == '2' or 'go' in choice_3.lower() or 'around' in choice_3.lower():
        print("""
        The door on the other side doesn't open. Did you really think it would 
        be that easy? Try something else. 
        """)
        input(prompt = '<Press Enter to continue> \n')
        chess_room()
        
    elif choice_3.lower() == '3' or 'play' in choice_3.lower():
        print("""
        That's bold but it looks like you don't have any other choice. You decide
        to act as a bishop while your partners take the roles of other missing
        pieces. You arrive to the point where your friend Ron sacrifices himself
        so you can checkmate the opposite king. However, you are utterly incompetent
        at chess. You remember the line he told you to go to (E), but not the 
        precise tile. You better guess quickly before your turn passes! A little
        hint, your current position is C5...
        
        1. Tile 7
        2. Tile 8
        3. Tile 3
        \n""")
        
        tile_nb = [3, 4, 7, 8]
        
#Creating a loop on a list with 'for' argument instead of while       
        for elements in tile_nb:
            try: #Try statement to cover non integer answer from the user
                tile = int(input(prompt = 'Your tile number \n\n')) #Defining the answer as integers              
                                                         
                if tile == 3:
                    print("""
                    You've put the opposite king in checkmate! You can access the
                    next door. You run towards it and don't even care about Ron 
                    because he's lucky enough to have Hermione taking care of him.
                    \n""")
                    input(prompt = '<Press Enter to continue> \n')
                    jedusor_room() #Sends the user to the next variable jedusor_room
                    break #Preventing the loop from going after finding the right answer
                               
                elif tile == 7 or tile == 1:
                    print("""
                    It's a cool move but totally useless. The game goes on but with
                    your expert Ron out, you wind up facing the opposite Queen... 
                    \n""")
                    fail() #Display fail statement informing the user of the game over
                    break #Preventing the loop from going after the fail statement 
                              
                elif tile == 8 or tile == 2:
                    print("""
                    You do suck at chess. If you knew how to play, you would have
                    noticed you couldn't go to E8 from C5 while being a Bishop.
                    You might be the chosen one but not at chess. You are sent 
                    back to your tile C5.
                    \n""")
                                   
                else:
                    print("""
                    That's not even in the numbers displayed. Hogwarts should 
                    give out math classes instead of potions.
                    Try again.
                    \n""")  

            except ValueError: #Error statement to prevent the code from breaking after a wrong type of answer
                print("""
                One more wrong entry and I Avada Kedavra you in the face.
                Try again.
                \n""")
                
                
    elif choice_3.lower() == '4' or 'far' in choice_3.lower() or 'head' in choice_3.lower() or 'back' in choice_3.lower():
        print("""
        That might have been the smartest decision right now. You walk out of 
        there safely. However, you are no longer Dumbledore's favorite and
        Gryffondor loses the cup. 
        """)
        fail_chess()
        exit(0) #Lets the user out of the game without breaking the code after a game over
        
    else:
        print("""
        Man, the sorting hat should have put you at Hufflepuff. Try again.
        \n""")
        input(prompt = '<Press Enter to continue>')
        chess_room()
        
def jedusor_room():
    print("""
    You finally enter the last room to find professor Quirrell standing in front
    of a mirror that shows everything you want. Appearing in the back of his 
    head is the famous You-Know-Who, the most powerful dark sorcerer ever. 
    They offer you to team up with them and become an invincible trio. For your
    final stage, what do you chose to do?    
    
    1. Draw your wand
    2. Run away
    3. Team up with them
    4. Use the mirror
    \n""")
    
    choice_5 = input(prompt = 'What do you do? \n\n')
    
    if choice_5.lower() == '1' or 'draw' in choice_5.lower() or 'wand' in choice_5.lower():
        print("""
        You draw your wand in an effort to fight them.
        \n""")
        input(prompt = '<Press Enter to continue> \n')
        fail_1()
        
    elif choice_5.lower() == '2' or 'run' in choice_5.lower():
        print("""
        A fire curtain suddenly appears and you cannot escape anymore. You have 
        to do something else!
        \n""")
        input(prompt = '<Press Enter to continue> \n')
        jedusor_room()
        
    elif choice_5.lower() == '3' or 'team' in choice_5.lower():
        print("""
        Fine choice right here! You secretly associate with them and switch houses
        to join slytherin. You later become the most powerful dark sorcerer of
        all time which is kinda cool.
        \n""")
        dark_sorcerer()
        
    elif choice_5.lower() == '4' or 'use' in choice_5.lower() or 'mirror' in choice_5.lower():
        print("""
        You visualize what you wishes through the mirror and it magically appears 
        next to you. What do you visualize?
        
        1. The sorcerer's stone
        2. Headmaster Dumbledore
        
        \n""")
        
        choice_6 = input(prompt = 'What do you visualize? \n\n')
        
#Deining nested conditional statement
        if choice_6.lower() == '1' or 'sorcerer' in choice_6.lower() or 'stone' in choice_6.lower():
            print("""
            As soon as the stone appears in your hand, professor Quirrell kills
            you and take the stone away.
            \n""")
            fail_2()
        
        elif choice_6.lower() == '2' or 'headmaster' in choice_6.lower() or 'dumbledore' in choice_6.lower():
            print("""
            Dumbledore magically appears to your side. Realizing the situation, 
            he handles professor Quirrell. However, He-Who-Must-Not-Be-Named 
            manages to escape.
            \n""")
            input(prompt = '<Press Enter to continue> \n')
            success()
            
        else:
            print("""
            What have you done?! That wasn't in the choices. What you typed
            was actually a spell to go back in time...
            \n""")
            input(prompt = '<Press Enter to continue> \n')
            jedusor_room()
            
    else:
        print("""\n 
        Wrong Entry. Hurry up, you have to do something or they will
        kill you.
        """)
        jedusor_room()
        
#Defining all fail and success functions for the whole game
def fail_chess():
    print('Game Over')

def fail():
    print("""\n 
    The Queen smashes your face with her sword, leaving you dead.
    \n""")
    print('Game Over')
    exit(0)
    
def fail_1():
    print("""\n
    Did you really think that you, a first year student, stood a chance again
    this two sorcerers? They promptly kill you and go back to their plans.
    \n""")
    print('Game Over')
                  
def fail_2():
    print("""\n
    You died giving You-Know-Who the opportunity of taking his powers back.
    \n""")
    print('Game Over')
           
def dark_sorcerer():
    print("""\n
    You win! You win in an unexpected way but still!
    \n""")
        
def success():
    print("""\n
    You go back to class as the hero who saved the day and the wizard world!
    \n""")
    print('You Win!')
        
start() #Launches the game
```
